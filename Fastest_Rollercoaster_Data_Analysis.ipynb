```python
# Asking a question: What locations have the fastest rollercoasters (min 10 rollercoasters)?
```

```python
df['Location'].value_counts()
```

```python
df_filtered = df.query('Location != "Other"').groupby('Location')['Speed_mph'].agg(['mean', 'count']).query('count >= 10').sort_values('mean')['mean'].copy()
ax = df_filtered.plot(kind='barh', figsize=(10,5), title='Average Coaster Speed by Location')
ax.set_xlabel('Average Coaster Speed (mph)')

plt.show()
```

```python
df_filtered
# this is a series
```

```python
df_filtered.index
```

```python
df_filtered.values
```

```python
df_filtered = pd.DataFrame({'Location': df_filtered.index, 'Mean_Speed': df_filtered.values})
df_filtered
```

```python
fastest_speed = df_filtered['Mean_Speed'].max()
fastest_location = df_filtered.loc[df_filtered['Mean_Speed'].idxmax(), 'Location']

print(f"The highest mean rollercoaster speed with at least 10 rollercoasters is {fastest_location} at {round(fastest_speed,2)} mph.")
```
